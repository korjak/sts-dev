```python
import pandas as pd
import csv
```

```python
msr_test = pd.read_csv('msr_paraphrase_test.txt', error_bad_lines=False, sep='\t', lineterminator='\r', quoting=csv.QUOTE_NONE)
msr_test['Quality']=msr_test['Quality'].replace("\n", None);
msr_test['Quality']=msr_test['Quality'].apply(lambda x: int(x))
msr_test['Quality']=msr_test['Quality'].apply(lambda x: 5 if x>0 else 1)
msr_test['Quality']=msr_test['Quality'].apply(lambda x: float(x))
msr_test_s=msr_test[['#1 String', '#2 String', 'Quality']]
msr_test_s.insert(0, 'index', None) 
msr_test_s.insert(1, 'genre', None) 
msr_test_s.insert(2, 'filename', None)
msr_test_s.insert(3, 'year', None)
msr_test_s.insert(4, 'old_index', None) 
msr_test_s.insert(5, 'source1', None)
msr_test_s.insert(6, 'source2', None)
msr_test_s.rename(columns = {'Quality':'score', '#1 String':'sentence1', '#2 String':'sentence2'}, inplace = True) 
msr_test_s
```

```python
msr_train = pd.read_csv('msr_paraphrase_train.txt', error_bad_lines=False, sep='\t', lineterminator='\r', quoting=csv.QUOTE_NONE)
msr_train['Quality']=msr_train['Quality'].replace("\n", None);
msr_train['Quality']=msr_train['Quality'].apply(lambda x: int(x))
msr_train['Quality']=msr_train['Quality'].apply(lambda x: 5 if x>0 else 1)
msr_train['Quality']=msr_train['Quality'].apply(lambda x: float(x))
msr_train_s=msr_train[['#1 String', '#2 String', 'Quality']]
#msr_test_s['index'] = msr_test_s.index
msr_train_s.insert(0, 'index', None) 
msr_train_s.insert(1, 'genre', None) 
msr_train_s.insert(2, 'filename', None)
msr_train_s.insert(3, 'year', None)
msr_train_s.insert(4, 'old_index', None) 
msr_train_s.insert(5, 'source1', None)
msr_train_s.insert(6, 'source2', None)
msr_train_s.rename(columns = {'Quality':'score', '#1 String':'sentence1', '#2 String':'sentence2'}, inplace = True) 
msr_train_s
```

```python
sts = pd.read_csv('sts-mt.csv', error_bad_lines=False, sep='\t', header=None, warn_bad_lines=False, quoting=csv.QUOTE_NONE)
sts=sts[[0,1,4,5,3]]
sts.rename(columns = { 0:'filename', 1:'year',4:'sentence1', 5:'sentence2', 3:'score' }, inplace = True) 
sts.insert(1,'genre', None)
sts.insert(4,'old_index', None) 
sts.insert(5,'source1', None) 
sts.insert(6,'source2', None) 
sts
```

```python
sts_other = pd.read_csv('sts-other.csv', error_bad_lines=False,warn_bad_lines=False, sep='\t', header=None)
sts_other=sts_other[[0,1,4,5,3]]
sts_other.rename(columns = { 0:'filename', 1:'year',4:'sentence1', 5:'sentence2', 3:'score' }, inplace = True) 
sts_other.insert(1,'genre', None)
sts_other.insert(4,'old_index', None) 
sts_other.insert(5,'source1', None) 
sts_other.insert(6,'source2', None) 
sts_other
```

```python
sts_train_s = pd.read_csv('sts_train.csv.txt', error_bad_lines=False, sep='\t', quoting=csv.QUOTE_NONE)
sts_train_s
```

```python
multinlitrain=pd.read_csv("multinli_1.0_train.txt", sep='\t', error_bad_lines=False, warn_bad_lines=False)
multinlitrain['label1']=multinlitrain['label1'].apply(lambda x: 0 if x=="neutral" else 1 if x=="contradiction" else 5 if x=="entailment" else 9)
multinlitrain['label1']=multinlitrain['label1'].apply(lambda x: float(x))
multinlitrain_s=multinlitrain[['sentence1', 'sentence2', 'label1']]
multinlitrain_s=multinlitrain_s[multinlitrain_s.label1!=9.0]
multinlitrain_s.insert(0, 'index', None) 
multinlitrain_s.insert(1, 'genre', None) 
multinlitrain_s.insert(2, 'filename', None)
multinlitrain_s.insert(3, 'year', None)
multinlitrain_s.insert(4, 'old_index', None) 
multinlitrain_s.insert(5, 'source1', None)
multinlitrain_s.insert(6, 'source2', None)
multinlitrain_s.rename(columns = {'label1':'score'}, inplace = True) 
multinlitrain_s
```

```python
multinlidev=pd.read_csv("multinli_1.0_dev_matched.txt", sep='\t', error_bad_lines=False, warn_bad_lines=False)
multinlidev['label1']=multinlidev['label1'].apply(lambda x: 0 if x=="neutral" else 1 if x=="contradiction" else 5 if x=="entailment" else 9)
multinlidev['label1']=multinlidev['label1'].apply(lambda x: float(x))
multinlidev_s=multinlidev[['sentence1', 'sentence2', 'label1']]
multinlidev_s=multinlidev_s[multinlidev_s.label1!=9.0]
multinlidev_s.insert(0, 'index', None) 
multinlidev_s.insert(1, 'genre', None) 
multinlidev_s.insert(2, 'filename', None)
multinlidev_s.insert(3, 'year', None)
multinlidev_s.insert(4, 'old_index', None) 
multinlidev_s.insert(5, 'source1', None)
multinlidev_s.insert(6, 'source2', None)
multinlidev_s.rename(columns = {'label1':'score'}, inplace = True) 
multinlidev_s
```

```python
frames = [msr_test_s, msr_train_s, sts_train_s, sts, sts_other, multinlitrain_s, multinlidev_s]
data = pd.concat(frames, ignore_index=True)
testd=data[data["sentence1"].str.contains("vaccine",  na=False)]
data=data[~data["sentence1"].str.contains("vaccine",  na=False)]
data=data[~data["score"].isna()]
data
```

```python
import numpy as np
train, dev, test = np.split(data.sample(frac=1, random_state=42), [int(.7*len(data)), int(.8*len(data))])
train.to_csv("trainbig.tsv", sep = '\t')
dev.to_csv("devbig.tsv", sep = '\t')
test.to_csv("testbig.tsv", sep = '\t')
```
